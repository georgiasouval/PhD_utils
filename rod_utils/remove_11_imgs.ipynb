```python
import os
import shutil

dir = '/home/souval_g_WMGDS.WMG.WARWICK.AC.UK/Desktop/00Train'
day_dir = '/home/souval_g_WMGDS.WMG.WARWICK.AC.UK/Desktop/00Train/day'
night_dir = '/home/souval_g_WMGDS.WMG.WARWICK.AC.UK/Desktop/00Train/night'

if not os.path.exists(day_dir):
    os.mkdir(day_dir)
if not os.path.exists(night_dir):
    os.mkdir(night_dir)

for filename in os.listdir(dir):
    full_name =  os.path.join(dir, filename)
    if filename.startswith('day'):
        shutil.move(full_name, day_dir)
    elif filename.startswith('night'):
        shutil.move(full_name, night_dir)
```

```python
import os
import shutil

day_dir = '/home/souval_g_WMGDS.WMG.WARWICK.AC.UK/Desktop/00Train/day'

original_dirs = [
    '/home/souval_g_WMGDS.WMG.WARWICK.AC.UK/Desktop/data_day/images/test',
    '/home/souval_g_WMGDS.WMG.WARWICK.AC.UK/Desktop/data_day/images/train',
    '/home/souval_g_WMGDS.WMG.WARWICK.AC.UK/Desktop/data_day/images/val'
]

new_dirs = ['train', 'test', 'val']
for new_dir in new_dirs:
    os.makedirs(os.path.join(day_dir, new_dir), exist_ok=True)

for dir in original_dirs:
    print(dir)
    basename = os.path.basename(dir)
    files = [os.path.splitext(filename)[0] + '.raw' for filename in os.listdir(dir)]    
    for filename in files:
        full_name = os.path.join(day_dir, filename)
        shutil.move(full_name, os.path.join(day_dir, basename))
        print(f'Moved {full_name} to {os.path.join(day_dir, basename)}')
```
